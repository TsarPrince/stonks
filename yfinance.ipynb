```python
import yfinance as yf
import pandas as pd
from datetime import datetime

def download_stock_data(tickers, start_date, end_date):
    # Download stock data from Yahoo Finance
    data = yf.download(tickers, start=start_date, end=end_date)
    
    # Save data to CSV files
    for ticker in tickers:
        ticker_data = data['Adj Close'][ticker]
        filename = f"{ticker}_{datetime.today().strftime('%Y-%m-%d')}.csv"
        ticker_data.to_csv(filename)
        print(f"Data for {ticker} saved as {filename}")
```

```python
# List of NSE stock tickers
tickers = ["RELIANCE.NS", "TCS.NS", "INFY.NS"]

# Define date range (optional)
start_date = "2024-01-01"
end_date = datetime.today().strftime('%Y-%m-%d')

# Download the stock data
download_stock_data(tickers, start_date, end_date)
```

```python
investments = {
  'BEL.NS': [ {'qty': 10, 'price': 293.70, 'date': '2024-05-23'}],
  'IRFC.NS': [ {'qty': 10, 'price': 186.20, 'date': '2024-05-23'}],
  'RVNL.NS': [ {'qty': 10, 'price': 372.15, 'date': '2024-05-23'}],
  'TATAPOWER.NS': [ {'qty': 5, 'price': 448.90, 'date': '2024-05-23'}],
  'ADANIPOWER.NS': [ {'qty': 5, 'price': 716.10, 'date': '2024-05-23'}],
  'HAL.NS': [ {'qty': 1, 'price': 5183.95, 'date': '2024-05-23'}, {'qty': 1, 'price': 4348.75, 'date': '2024-09-20'}],
  'ULTRACEMCO.NS': [ {'qty': 1, 'price': 11530, 'date': '2024-07-11'}],
  'HINDALCO.NS': [ {'qty': 5, 'price': 694.65, 'date': '2024-07-11'}],
  'NTPC.NS': [ {'qty': 10, 'price': 376.30, 'date': '2024-07-11'}],
  'COALINDIA.NS': [ {'qty': 10, 'price': 502, 'date': '2024-07-11'}],
  'RECLTD.NS': [ {'qty': 10, 'price': 643.95, 'date': '2024-07-11'}, {'qty': 10, 'price': 538.30, 'date': '2024-09-20'}],
  'ONGC.NS': [ {'qty': 10, 'price': 305.50, 'date': '2024-07-11'}],
}
```

```python
total_inv = {}
for key, value in investments.items():
    total_inv[key] = sum([entry['price'] * entry['qty'] for entry in value])

total_inv = dict(sorted(total_inv.items(), key=lambda item: item[1]))
total_inv
```

```python
import numpy as np
import pandas as pd

# Convert the date string in the investments dictionary to datetime for easy comparison
for ticker, entries in investments.items():
    for entry in entries:
        entry['date'] = pd.to_datetime(entry['date']).tz_localize('UTC')

investments
```

```python
# Iterate over the tickers and calculate the inv_amt and inv_qty
for ticker in investments.keys():
    
    # Initialize inv_amt and inv_qty columns with NaN
    df[('inv_amt', ticker)] = np.nan
    df[('inv_qty', ticker)] = np.nan
    
    # Loop over each date in the DataFrame
    for i, date in enumerate(df.index):
        
        # Initialize variables to accumulate total amount and total quantity
        total_amt = 0
        total_qty = 0
        
        # Iterate over investment entries for the ticker
        for entry in investments[ticker]:
            # If the date is greater than or equal to the investment date, add the qty and amt
            if date >= entry['date']:
                total_qty += entry['qty']
                total_amt += entry['qty'] * entry['price']
        
        # Only set values if investments have been made, otherwise leave it as NaN
        if total_qty > 0:
            # Calculate the average amount till the current date
            avg_amt = total_amt / total_qty
            
            # Set the inv_amt and inv_qty for the current date
            df.loc[date, ('inv_amt', ticker)] = avg_amt
            df.loc[date, ('inv_qty', ticker)] = total_qty
```

```python
import matplotlib.pyplot as plt

# Get the last available date in the DataFrame
last_day = df.index[-1]

# Extract 'pl' and 'inv_amt' for the last day for all tickers
pl_last_day = [df[('pl', ticker)][last_day] for ticker in tickers]
inv_amt_last_day = [df[('inv_amt', ticker)][last_day] * df[('inv_qty', ticker)][last_day] for ticker in tickers]

# Sort tickers based on increasing order of inv_amt_last_day
sorted_indices = np.argsort(inv_amt_last_day)
tickers = np.array(tickers)[sorted_indices]
inv_amt_last_day = np.array(inv_amt_last_day)[sorted_indices]
pl_last_day = np.array(pl_last_day)[sorted_indices]

# Plotting
fig, ax = plt.subplots()

# Bar width
bar_width = 0.4

# Create a bar graph for inv_amt (base bar)
ax.bar(tickers, inv_amt_last_day, bar_width, label='Investment Amount')

# Stack pl on top of inv_amt
ax.bar(tickers, pl_last_day, bar_width, label='Profit/Loss')

# Add labels and title
ax.set_xlabel('Tickers')
ax.set_ylabel('Amount')
ax.set_title('Investment Amount and Profit/Loss on Last Day')
plt.xticks(rotation=90)

# Add a legend
ax.legend()
ax.axhline(0, color='black',linewidth=0.8)  # Draw the x-axis at y=0
```
